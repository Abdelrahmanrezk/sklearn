```python
from sklearn.model_selection import KFold
```

```python
'''
    KFold used to split your data to some flods as you need
    random_state used because the same data will be floded if use other algorithm
    This help you to train and test alot because its use some of data in train then in next split use it as test
'''
X = np.arange(120).reshape(60, 2)
y = np.arange(60).reshape(60,1)
X.shape
```

```python
kf = KFold(n_splits=4, random_state=44, shuffle =True)
```

```python

for train_index,test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print('X_train \n' , X_train)
    print('X_test \n' , X_test)
    print('y_train \n' ,y_train)
    print('y_test \n' , y_test)
```
